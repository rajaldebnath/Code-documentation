## Jellyfish counting kmers and estimating genome size and coverage of sequencing/ depth of sequencing

```python
!ls
```

```python
# install jellyfish using homebrew by tapping into brewsci/bio from github and than install jellyfish
# we will be using the spades corrected reads which were used for assembling the genome
!mkdir jellyfish_gcoverage_gsize
!jellyfish count -t 8 -C -m 21 -s 5G -o jellyfish_gcoverage_gsize/19mer_out.jf Assembly/corrected/RJ12_R1.paired.fastq.00.0_0.cor.fastq Assembly/corrected/RJ12_R2.paired.fastq.00.0_0.cor.fastq
!jellyfish histo jellyfish_gcoverage_gsize/19mer_out.jf -o jellyfish_gcoverage_gsize/19mer_out.histo
!cat jellyfish_gcoverage_gsize/19mer_out.histo
```

## We will use R to plot the histogram and see the coverage maxima, for this we will call R within jupyter notebook, for this we first have to install rpy2 package and load it into ipython with the below comand

```python
# install rpy2 using pip, and then load it into the jupyter environment

%reload_ext rpy2.ipython
```

```python
%%R #magic cell for R
library(tidyverse)
kmer_histo <- read.table("jellyfish_gcoverage_gsize/19mer_out.histo")

kmer_histo #there are a total of 617 rows in the dataframe
```

```python
%%R -w 4 -h 3 --units in -r 120
library(tidyverse)

#Since the maximum number of distinct kmers were observed at a frequency of 55, lets generate the plot 
kmer_histo_plot <- ggplot(kmer_histo[2:100,], aes(x = V1, y = V2)) +
                    geom_line() +
                    theme_classic() +
                    xlab("Times a kmer is observed") + ylab("No of kmers observed \nat a given frequency") +
                    geom_vline(xintercept = 55, linetype="dotted", color = "blue", size=0.5)
kmer_histo_plot
```

## Estimating genome size of EBRJ12

**The first datapoint in the histo file is the number of times a kmer is observed which is 1, this is not a single copy region, to get the single copy region we need to analyze from the 2nd to the last datapoint i.e, 617 and calculate total kmers in the distribution**

```python
%%R
#Calculate the estimated genome size after obtaining the coverage of sequencing (peak size = 55)

#total number of kmers for single copy region
print(sum(as.numeric(kmer_histo[2:617,1] * kmer_histo[2:617,2])))

#total number of kmers divided by coverage of sequencing
estimated_genome_size_EBRJ12 <-  sum(as.numeric(kmer_histo[2:617,1] * kmer_histo[2:617,2]))/55
print(estimated_genome_size_EBRJ12/10^6) #in megabasepairs
```
